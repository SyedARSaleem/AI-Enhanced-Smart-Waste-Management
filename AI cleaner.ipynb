```python
# AI Coursework 1
# Date of Last Update: 20/02/2024
# Name: Syed Abdul Rafey Saleem


from tabulate import tabulate


# --------------------- Function to highlight the current position of the cleaner

def print_live_table(current, position):
    highlighted = [row.copy() for row in current]
    highlighted[position[0]][position[1]] = f"\033[1;31m{highlighted[position[0]][position[1]]}\033[0m"  
    print(tabulate(highlighted, tablefmt='fancy_grid'))



# -------------- AI Cleaner working --------------
    
def clean(current, position):
    print("\n------------ Cleaning Location (" , position[0] + 1 , ", " , position[1] + 1 , ") .........")
    print_live_table(current, position)
    current[position[0]][position[1]] = "CLEAN"
    print("\n------------ Location (" , position[0] + 1 , "," , position[1] + 1 , ") Cleaned")
    return current

def turnleft(position):
    print_live_table(current, position)
    print("\n\n Currently at position      (" , position[0] + 1 , "," , position[1] + 1 , ")")
    position[1] = position[1] - 1
    print("\n------------ Turning left to Location (" , position[0] + 1 , "," , position[1] + 1 , ")  ............")
    return position

def turnright(position):
    print_live_table(current, position)
    print("\n\n Currently at position      (" , position[0] + 1 , "," , position[1] + 1 , ")")
    position[1] = position[1] + 1
    print("\n------------ Turning right to Location (" , position[0] + 1 , "," , position[1] + 1 , ")")
    return position   
    
def moveup(position):
    print_live_table(current, position)
    print("\n\n Currently at position      (" , position[0] + 1 , "," , position[1] + 1 , ")")
    position[0] = position[0] - 1
    print("\n------------ Moving up to Location (" , position[0] + 1 , "," , position[1] + 1 , ")")
    return position    
    
def movedown(position):
    print_live_table(current, position)
    print("\n\n Currently at position      (" , position[0] + 1 , "," , position[1] + 1 , ")")
    position[0] = position[0] + 1
    print("\n------------ Moving down to Location (" , position[0] + 1 , "," , position[1] + 1 , ")")
    return position
  

# The cleaner detects the nearest waste location first and then goes forward to clean that location
# It repeats this process until the goal state is achieved that is all locations cleaned
def action(current, goal, dirty_loc, position):
    shortest = 11
    closest = [0,0]
    for i in range(len(dirty_loc)):
        distance = abs(dirty_loc[i][0] - position[0]) + abs(dirty_loc[i][1] - position[1])
        if distance < shortest:
            shortest = distance
            closest[0] = dirty_loc[i][0]
            closest[1] = dirty_loc[i][1]
            
    while True:
        x = closest[0] - position[0]
        y = closest[1] - position[1]

        if y < 0:
            position = turnleft(position)
            continue
        elif y > 0:
            position = turnright(position)
            continue
        elif x < 0:
            position = moveup(position)
            continue
        elif x > 0:
            position = movedown(position)
            continue
        else:
            current = clean(current, position)
            break
    
    dirty_loc.remove(closest)
    
    if current == goal:
        print_live_table(current, position)
        print("\n\n\n=================  All The Waste has been collected and disposed in an eco-friendly manner  ================\n\n")
        return current
    else:
        return action(current, goal, dirty_loc, position)


    

# -------- Validating the size of the environment

def check_range():
    size = int_only()
    while (size < 1) or (size > 5):
        print("Value entered is out of range! please try again\n")
        size = int_only() 
    return size
    

    
# -------- Validating the type of the size    

def int_only():
    size = 0
    try:
        size = int(input("Can you please enter the size 'n' of the n x n environment:   "))
        return size
    except:
        print("Please enter an integer value only\n")
        return int_only()

    

# ----------- Asking User for coordinates for dirty places and starting point

def input_coordinates(size):
    coordinates = input("Please enter the coordinates of the location \nPlease enter as 'r, c' where r is the row and c is the column of the location\n      ")
    coordinates = coordinates.strip()
    try:                              # Validating the format of the coordinates input
        row = int(coordinates[0])
    except:
        print("\nInvalid format! \n")
        return input_coordinates(size)
    try: 
        column = int(coordinates[-1])
    except:
        print("\nInvalid format! \n")
        return input_coordinates(size)
    
    if (row < 1) or (row > size):      # Validating the range of the coordinates input
        print("\nRow coordinate out of range\nPlease enter a valid row number\n")
        return input_coordinates(size)
    
    if (column < 1) or (column > size):
        print("\nColumn coordinate out of range\nPlease enter a valid column number\n")
        return input_coordinates(size)
    
    return row, column



# ----------------- Inputing Environment size

print("\n\n   --------------------  ENVIRONMENT SIZE  ----------------------\n")
size = check_range()



# ---------------- Creating an environment according to the user's input

current = [["CLEAN"] * size for i in range(size)]
goal = [["CLEAN"] * size for i in range(size)]



# ------------------ Asking User for dirty locations

print("\n\n   --------------  WASTE LOCATION COORDINATES  --------------\n")
end = 'Y'
count = 0
l = size * size
dirty_loc = []
while end != 'N':
    r, c = input_coordinates(size)
    dirty_loc.append([])
    dirty_loc[count].append(r-1)
    dirty_loc[count].append(c-1)
    count = count + 1
    end = input("\nDo you want to enter more waste locations?\nEnter 'Y' for Yes and 'N' for No      ")
    end = end.upper()

    
# ------------------- Updating dirty loactions
print(dirty_loc)
for x in range(len(dirty_loc)):
    current[dirty_loc[x][0]][dirty_loc[x][1]] = "DIRTY"

 

 
# ------------------- Asking user for the intitial location of the cleaner  

print("\n\n   --------------  INITIAL LOCATION COORDINATES  --------------")           
r, c = input_coordinates(size)
position = []
position.append(r-1)
position.append(c-1)



# -------------------- AI Robot Cleaner is going to take actions according to the environment 

print("\n\n")
print_live_table(current, position)
print("\n\n\n\n -----------------------   AI MODE ON   ------------------------- \n\n")
#print(tabulate((current), tablefmt = 'fancy_grid'))
final = action(current, goal, dirty_loc, position)
print(tabulate((final), tablefmt = 'fancy_grid'))
```
